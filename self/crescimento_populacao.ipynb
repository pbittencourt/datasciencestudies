# Crescimento da população brasileira (1980-2016)

Dados obtidos de DataSus (fazer referência futura)

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('populacao_brasileira.csv', sep=';')
df.columns = ['Ano', 'População']
df.head()

ano = df['Ano']
pop = df['População']

plt.plot(ano, pop, color='#333333', linestyle=':')
plt.bar(ano, pop, color='#cccccc')
plt.title('Crescimento da população brasileira (1980-2016)')
plt.xlabel('Ano')
plt.ylabel('População')
#plt.show()
plt.savefig('pop.png', dpi=300)  # não funciona junto de show() (a imagem vem em branco)
```
